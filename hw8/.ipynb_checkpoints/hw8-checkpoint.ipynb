```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.load('x.npy')
y = np.load('y.npy')
```

```python
plt.scatter(x, y)
```

```python
def func(x, c):
    return c[0]*(x**2) + c[1]*x + c[2]
```

```python
def count_inliers(x, c, y):
    count = 0
#     print('\nC=',c)
    for i in range(len(y)):
        if math.sqrt((func(x[i], c) - y[i])**2) <= 0.05:
#             print('hit!')
            count += 1
    return count
```

```python
def ransac(x, y):
    n = 1000
    mscore = 0
    c = np.zeros(3)
    for _ in range(n):
        tmpi = np.random.choice(np.arange(120), 3)
        tmpx = [[x[tmpi[0]]**2, x[tmpi[0]], 1],
                [x[tmpi[1]]**2, x[tmpi[1]], 1],
                [x[tmpi[2]]**2, x[tmpi[2]], 1]]
        tmpy = [y[tmpi[0]], y[tmpi[1]], y[tmpi[2]]]
        tmpc = np.linalg.lstsq(tmpx, tmpy)[0]
        tmpscore = count_inliers(x, tmpc, y)
        if tmpscore > mscore:
            print('C=', c, ' score=', tmpscore)
            c = tmpc
            mscore = tmpscore
            
    return c
    
c = ransac(x, y)
```

```python
plt.plot(x, y, 'o', label='Original data')
plt.plot(x, func(x, c), '*', label='Fitted line')
plt.legend()
plt.show()
```
