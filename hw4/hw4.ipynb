```python
import numpy as np
from skimage.transform import pyramid_gaussian
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open('test-image-MSRA-3_97_97769.jpg')
img.load()
data = np.asarray(img, dtype=np.float)
```

```python
gs = np.average(data, axis=2)
```

```python
plt.imshow(gs, cmap='gray')
```

```python
gpl1 = np.asarray(tuple(pyramid_gaussian(gs, max_layer=4, sigma=3)))
gpl2 = np.asarray(tuple(pyramid_gaussian(gs, max_layer=4, sigma=9)))
```

```python
on_off = gpl1 - gpl2
off_on = gpl2 - gpl1
```

# ON-OFF

```python
fig = plt.figure(figsize=(20,10))
for i in range(4):
    fig.add_subplot(2, 2, i+1)
    plt.imshow(on_off[i+1], cmap='gray')
plt.show()
```

# OFF-ON

```python
fig = plt.figure(figsize=(20,10))
for i in range(4):
    fig.add_subplot(2, 2, i+1)
    plt.imshow(off_on[i+1], cmap='gray')
plt.show()
```
